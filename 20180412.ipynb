```python
def stemming(nameList):
    names = nameList
    stems=[]
    for name in names:
        if '有限公司' or '股份有限公司' in name:
            stem=[]
            for i,w in enumerate(list(name)):
                if w!='股'and w!='份'and w!='有'and w!='限'and w!='公'and w!='司' and w!='(' and w!=')':
                    stem.append(w)
        stems.append(stem)
    return stems
```
